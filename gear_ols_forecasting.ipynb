```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
```

```python
# Example historical data
data = {
        'Year': [2019, 2020, 2021, 2022, 2023],
        'Revenue': [1000000, 1100000, 1200000, 1300000, 1400000],
        'Net Income': [100000, 105000, 110000, 115000, 120000],
        'Total Debt': [500000, 550000, 600000, 650000, 700000],
        'Total Equity': [500000, 600000, 700000, 800000, 900000]
}

df = pd.DataFrame(data)
```

```python
# Calculate gearing ratio (Debt/Equity)
df['Gearing Ratio'] = df['Total Debt'] / df['Total Equity']
# Define features and target variable
X = df[['Year', 'Gearing Ratio']]
y = df['Net Income']
# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)
# Predict future values
future_years = np.array([[2024, df['Gearing Ratio'].iloc[-1]], [2025, df['Gearing Ratio'].iloc[-1]]])
predictions = model.predict(future_years)
# Create a DataFrame for the predictions
future_df = pd.DataFrame({
    'Year': [2024, 2025],
    'Predicted Net Income': predictions
    })

print(future_df)
```

```python
# Plot the historical and predicted net income
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Net Income'], label='Historical Net Income', marker='o')
plt.plot(future_df['Year'], future_df['Predicted Net Income'], label='Predicted Net Income', marker='x')
plt.xlabel('Year')
plt.ylabel('Net Income')
plt.title('Historical and Predicted Net Income')
plt.legend()
plt.grid(True)
plt.show()
```
